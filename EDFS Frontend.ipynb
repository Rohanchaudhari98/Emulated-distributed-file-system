## EDFS Commands

```python
command = input("What command do you want to execute: ")
if command == "put":
    file = widgets.FileUpload(
        accept='',  
        multiple=False  
        )
    display(file)
```

```python
flg = 1
putflag = 0
while True:
    if flg != 1:
        command = input("What command do you want to execute: ")
        if command == "put":
            putflag == 1
            print("Go to above cell")
            putflag = 0
            break
    if command == "mkdir" or command == "ls":
        name = input("Enter the directory name: ")
        
    elif command == "cat" or command == "rm" or command == "getPartitionLocations":
        name = input("Enter the file name: ")
        
    elif command == "put" and putflag == 0:
        putflag = 1
        print(file.value.keys())
        name = list(file.value.keys())[0]
        dirname = input("Enter the directory to upload to: ")
        numpart = input("Enter the number of partitions: ")
        url = "http://192.168.1.37:80/"+command+"/"+name+dirname+"/"+numpart
        r = requests.put(url)
        print(r.text)
        
    elif command == "readPartition":
        name = input("Enter the file name: ")
        partnum = input("Enter the partition number to read: ")

    if command == "mkdir":  
        url = "http://192.168.1.37:80/"+command+"/"+name[1:len(name)]
        r = requests.post(url)
        print(r.text)

    elif command == "ls":
        url = "http://192.168.1.37:80/"+command+"/"+name[1:len(name)]
        r = requests.get(url)
        print(r.json())

    elif command == "cat":
        url = "http://192.168.1.37:80/"+command+"/"+name
        r = requests.get(url)
        print()
        print(r.text)

    elif command == "rm":
        url = "http://192.168.1.37:80/"+command+"/"+name[1:len(name)]
        r = requests.delete(url)
        print(r.text)

    elif command == "getPartitionLocations":
        url = "http://192.168.1.37:80/"+command+"/"+name
        r = requests.get(url)
        print(r.text)

    elif command == "readPartition":
        url = "http://192.168.1.37:80/"+command+"/"+name+"/"+partnum
        r = requests.get(url)
        print(r.text)
    print()
    
    flag = input("Do you want to continue: ")
    if flag == "yes":
        flg = 0
        continue
    else:
        print("End of EDFS operations")
        break
```
